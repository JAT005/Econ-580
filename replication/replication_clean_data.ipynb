# Extension (crowding out)

```python
gov_fin_df = pd.read_excel('govt_borrow.xlsx')
gov_fin_df = gov_fin_df.iloc[:-2]
gov_fin_df
gov_fin_df = gov_fin_df.rename(columns={
    gov_fin_df.columns[0]: 'country'
})
gov_long = gov_fin_df.melt(
    id_vars='country',          # stays fixed
    var_name='year',            # new column name
    value_name='debt_pct_gdp'   # values column
)
gov_long
gov_long = gov_long.replace("no data", None)
gov_long['debt_pct_gdp'] = pd.to_numeric(gov_long['debt_pct_gdp'], errors='coerce')

gov_long
```

```python
gov_long = gov_long[gov_long['country'].isin(countries)]
gov_long = gov_long[gov_long['year'].isin([1989, 1990, 1991])]
gov_long = gov_long[['country', 'debt_pct_gdp']]
gov_long = gov_long.groupby('country').mean('debt_pct_gdp').reset_index()
gov_long
```

```python
saving_df = pd.read_csv('gross_savings.csv')
saving_df
```

```python
saving_long = saving_df.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    var_name='year',
    value_name='savings (% of GDP)'
)
saving_long
saving_long['year'] = pd.to_numeric(saving_long['year'], errors='coerce')
saving_long = saving_long.rename(columns={
    saving_long.columns[0]: 'country'
})

saving_long
```

```python
saving_long['country'].unique()
```

```python
saving_long['year']
```

```python
saving_long = saving_long[saving_long['country'].isin(countries)]
saving_long = saving_long[saving_long['year'].isin([1989, 1990, 1991])]
saving_long = saving_long[['country', 'savings (% of GDP)']]
saving_long = saving_long.groupby('country').mean('savings (% of GDP)').reset_index()
saving_long
```

```python
crowdout_df = saving_long.merge(gov_long, how = 'left')
crowdout_df['govsav'] = crowdout_df['debt_pct_gdp']/crowdout_df['savings (% of GDP)']
crowdout_df
```
